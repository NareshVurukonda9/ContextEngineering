# Exercise: Multi-Format Document Ingestion Pipeline

## Objective
Build a document ingestion pipeline that can load documents from multiple file formats (JSON, Markdown, PDF), normalize them into a consistent format, and combine them into a single collection.

## What You'll Do
This exercise contains **5 TODOs** organized into a document ingestion pipeline:

### The 5 TODOs

| Todo | Component | Task |
|------|-----------|------|
| **TODO 1** | JSON Loading | Load documents from a JSON file using LangChain's JSONLoader |
| **TODO 2** | Markdown Loading | Load documents from a Markdown file using TextLoader |
| **TODO 3** | PDF Loading | Load documents from a PDF file using PyPDFLoader |
| **TODO 4** | Normalization | Standardize all documents (clean text, add metadata) regardless of source |
| **TODO 5** | Pipeline | Combine all loaders into one complete ingestion function |

## How the Pipeline Works

1. **Load**: Each TODO (1-3) loads documents from a different file format
2. **Normalize**: TODO 4 ensures all documents have the same structure and metadata
3. **Combine**: TODO 5 runs all loaders, normalizes their output, and merges everything together

```
JSON File → Load → Normalize ──┐
                                ├→ Combined Documents
Markdown File → Load → Normalize┤
                                ├→ [Document, Document, ...]
PDF File → Load → Normalize ───┘
```

## What Each Function Should Do

- **`load_json()`**: Use JSONLoader with the provided schema to extract documents from JSON
- **`load_markdown()`**: Use TextLoader to read markdown files as documents
- **`load_pdf()`**: Use PyPDFLoader to extract pages from PDF files
- **`normalize_documents()`**: 
  - Clean/strip whitespace from document content
  - Add metadata tagging (source_type, original metadata)
  - Return standardized Document objects
- **`ingest_all()`**: Call all three loaders, normalize their outputs, and return combined list

## Key Concepts You'll Learn

- **Document Loaders**: Different tools for different file formats
- **Metadata**: Tracking where documents came from (source_type)
- **Normalization**: Making data consistent across different sources
- **Pipeline**: Chaining multiple steps together into one reusable function

## Expected Output

When the pipeline runs, it should print metadata and content samples from the first 5 documents, showing documents loaded from all three formats (JSON, Markdown, PDF).

````python
from langchain_community.document_loaders import JSONLoader, PyPDFLoader, TextLoader
from langchain_core.documents import Document
import json


# -------------------------
# TODO 1: Load JSON
# -------------------------
def load_json(file_path):

    loader = JSONLoader(
        file_path=file_path,
        jq_schema=".[]",   # assumes list of objects
        text_content=False
    )

    docs = loader.load()

    return docs


# -------------------------
# TODO 2: Load Markdown
# -------------------------
def load_markdown(file_path):

    loader = TextLoader(file_path)

    docs = loader.load()

    return docs


# -------------------------
# TODO 3: Load PDF
# -------------------------
def load_pdf(file_path):

    loader = PyPDFLoader(file_path)

    docs = loader.load()

    return docs


# -------------------------
# TODO 4: Normalize Documents
# -------------------------
def normalize_documents(docs, source_type):

    normalized = []

    for doc in docs:

        content = doc.page_content

        # TODO: clean text
        content = content.strip()

        # TODO: attach metadata
        metadata = doc.metadata or {}
        metadata["source_type"] = source_type

        normalized.append(
            Document(
                page_content=content,
                metadata=metadata
            )
        )

    return normalized


# -------------------------
# TODO 5: Combine All Sources
# -------------------------
def ingest_all():

    all_docs = []

    # JSON
    json_docs = load_json("data/data.json")
    all_docs.extend(normalize_documents(json_docs, "json"))

    # Markdown
    md_docs = load_markdown("data/notes.md")
    all_docs.extend(normalize_documents(md_docs, "markdown"))

    # PDF
    pdf_docs = load_pdf("data/report.pdf")
    all_docs.extend(normalize_documents(pdf_docs, "pdf"))

    return all_docs


# -------------------------
# Run Pipeline
# -------------------------
documents = ingest_all()

for doc in documents[:5]:
    print(doc.metadata, "\n", doc.page_content[:100], "\n")
````
